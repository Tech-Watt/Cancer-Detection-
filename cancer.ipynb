Creating the Machine learning model

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import numpy as np
```

Separating data into features and targets and Spliting into training and testing set

```python
features = final_data.drop('Level', axis=1)
target_data = final_data['Level']

target_data = np.array([ 0 if x == 'Low' else 1 if x == 'Medium' 
                        else 2 for x in target_data])

X_train,X_test,Y_train,Y_test = train_test_split(features,target_data,
        random_state=42,test_size=0.20)
```

LOW = 0
MEDIUM = 1
HIGH = 2

Building a preprocessing,model pipeline and training the model.

```python
pipeline = Pipeline([
    ('scaler',StandardScaler()),
    # ('model',LinearRegression()),
    ('randomforest',RandomForestClassifier(n_estimators=100)),
])

pipeline.fit(X_train,Y_train)
```

Evaluating the trained model on the Test dataset

```python
y_pred = pipeline.predict(X_test)
mae = mean_absolute_error(Y_test, y_pred)
mse = mean_squared_error(Y_test, y_pred)
print(f'Mean absolute error: {mae}')
print(f'Mean squared error: {mse}')
```

Visualizing the output

```python
# Create a DataFrame with the true and predicted values
data = {'True Values': Y_test, 'Predicted Values': y_pred}
df = pd.DataFrame(data)

# Create the scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Predicted Values', y='True Values', data=df)
plt.plot([df['True Values'].min(), df['True Values'].max()], [df['True Values'].min(), df['True Values'].max()], 'r--', lw=2)
plt.xlabel('Predicted Values')
plt.ylabel('True Values')
plt.title('Scatter Plot of True vs Predicted Values')
plt.show()
```
